```python
#Bài tập về nhà bài 4.1 - Mô tả ứng dụng và cài đặt demo ứng dụng với 2 thuật toán trên.

import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

X_khachang = np.array([
    [2, 5], [3, 7], [1, 4], [10, 50], [12, 55],
    [11, 48], [5, 20], [6, 22], [5, 18]
])

model_km = KMeans(n_clusters=3, random_state=42)
y_clusters = model_km.fit_predict(X_khachang)

plt.scatter(X_khachang[:, 0], X_khachang[:, 1], c=y_clusters, cmap='viridis',s=100)
plt.scatter(model_km.cluster_centers_[:, 0], model_km.cluster_centers_[:, 1], c='red', s=200, marker='X', label='Tâm cụm')
plt.title("Phân nhóm khách hàng bằng K-Means")
plt.xlabel("Số lần mua")
plt.ylabel("Số tiền chi tiêu (Triệu VND)")
plt.legend()
plt.show()
```

```python
#Bài tập về nhà bài 4.2 - Mô tả ứng dụng và cài đặt demo ứng dụng với 2 thuật toán trên.

from sklearn.neighbors import KNeighborsClassifier

X_train = np.array([[10, 10], [11, 11], [10.5, 10.2], [5, 5], [4, 6], [5.5, 4.5]])
y_train = np.array([1, 1, 1, 0, 0, 0])

X_new = np.array([[6, 6]])

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
ket_qua = knn.predict(X_new)

plt.scatter(X_train[y_train==1, 0], X_train[y_train==1, 1], color='blue', label='Đạt')
plt.scatter(X_train[y_train==0,0], X_train[y_train==0,1], color='red', label='Lỗi')
plt.scatter(X_new[:, 0], X_new[:, 1], marker='*',color='green', s=200, label='Sản phẩm mới')
plt.title(f"Phân loại K-NN: Sản phẩm mới dự báo là  {'Đạt' if ket_qua[0]==1 else 'Lỗi'}")
plt.legend()
plt.show()
```

```python
#Bài tập về nhà bài 5 - Mô tả ứng dụng và cài đặt demo ứng dụng với các thuật toán được học trên lớp lý thuyết.

from sklearn.linear_model import LinearRegression

X_dientich = np.array([[30], [45], [60], [80], [100], [120]])
y_gia = np.array([1.5,2.3,3.1,4.5,5.2,6.8])

reg = LinearRegression()
reg.fit(X_dientich, y_gia)

dientich_moi = np.array([[75]])
gia_du_bao = reg.predict(dientich_moi)

plt.scatter(X_train[y_train==1, 0], X_train[y_train==1, 1], color='blue', label='Đạt')
plt.plot(X_dientich, reg.predict(X_dientich), color='red', label='Đường dự báo')
plt.scatter(dientich_moi, gia_du_bao, marker='s',color='green', s=100, label='Dự báo 75m2')
plt.title(f"Hồi quy tuyến tính: Nhà 75m2 giá khoảng {gia_du_bao[0]:.2f} Tỷ")
plt.xlabel("Diện tích(m2)")
plt.ylabel("Giá(Tỷ VND)")
plt.legend()
plt.show()
```
