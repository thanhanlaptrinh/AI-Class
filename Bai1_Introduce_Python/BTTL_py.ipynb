```python
#Câu 13: Viết hàm trộn 2 mảng một chiều thành 1 mảng một chiều với mỗi phần tử của mảng mới là tổng của 2 phần tương ứng từ 2 mảng cho trước.
#Trong quá trình trộn 2 mảng nếu mảng nào còn phần tử thì các phần tử còn lại của mảng đó sẽ đưa vào mảng mới.
#Ví dụ:
#Mảng a: 3 9 1 4
#Mảng b: 2 7 4 3 2 8
#Mảng kết quả: 5 16 5 7 2 8 23.
def tron_mang(a, b):
  mang_moi = []
  min_len = min(len(a), len(b))

  for i in range(min_len):
    mang_moi.append(a[i] + b[i])

  if len(a) > len(b):
    mang_moi.extend(a[min_len:])
  else:
    mang_moi.extend(b[min_len:])
  return mang_moi

a = [3, 9, 1, 4]
b = [2, 7, 4, 3, 2, 8]
kq = tron_mang(a, b)
print(f"Mảng a: {a}")
print(f"Mảng b: {b}")
print(f"Kết quả sau khi bánh tráng trộn: {kq}")
```

```python
import random

# d. Tìm dòng có tổng lớn nhất.
def max_row_sum(a):
    max_sum = sum(a[0])
    index = 0
    for i in range(1, len(a)):
        current_sum = sum(a[i])
        if current_sum > max_sum:
            max_sum = current_sum
            index = i
    print(f"Dòng có tổng lớn nhất là {index}: Tổng = {max_sum}")

# e. Tìm cột có tích nhỏ nhất.
def min_product_column(a):
    min_product = float('inf')
    index = 0
    for j in range(len(a[0])):
        product = 1
        for i in range(len(a)):
            product *= a[i][j]
        if product < min_product:
            min_product = product
            index = j
    print(f"Cột có tích nhỏ nhất là {index}: Tích = {min_product}")

# f. Xuất ra các phần tử thuộc dòng chẵn và cột lẻ trong a.
def print_ptu_dchan_cle(a, m, n):
    found = False
    for i in range(0, m, 2):
        for j in range(1, n, 2):
            print(f"a[{i}][{j}] = {a[i][j]}")
            found = True
    if not found:
        print("Không có phần tử nào thỏa mãn.")

# g. Tính trung bình cộng các phần tử chẵn thuộc dòng lẻ của a.
def avg_ptu_chan_dle(a, m, n):
    val = []
    for i in range(1, m, 2):
        for j in range(n):
            if a[i][j] % 2 == 0:
                val.append(a[i][j])
    if len(val) > 0:
        print(f"Trung bình cộng các phần tử chẵn thuộc dòng lẻ là: {sum(val)/len(val):.2f}")
    else:
        print("Không có phần tử chẵn ở dòng lẻ.")

# h. Tính trung bình cộng các phần tử thuộc biên.
def avg_ptu_bien(a, m, n):
    bien = []
    for i in range(m):
        for j in range(n):
            if i == 0 or i == m - 1 or j == 0 or j == n - 1:
                bien.append(a[i][j])

    if bien:
        print(f"Trung bình cộng các phần tử thuộc biên là: {sum(bien)/len(bien):.2f}")
    else:
        print("Không có phần tử biên.")

# i. Tính trung bình tích các phần tử không thuộc biên.
def avg_tich_kbien(a, m, n):
    kbien = []
    for i in range(m):
        for j in range(n):
            # Không thuộc biên tức là không phải dòng/cột đầu và cuối
            if i != 0 and i != m - 1 and j != 0 and j != n - 1:
                kbien.append(a[i][j])

    if kbien:
        tich = 1
        for x in kbien:
            tich *= x
        print(f"Trung bình tích các phần tử không thuộc biên là: {tich/len(kbien):.2f}")
    else:
        print("Không có phần tử không thuộc biên (mảng quá nhỏ).")

def menu():
    print("\n- MENU -")
    print("1. Tạo mảng a chứa các số nguyên ngẫu nhiên.")
    print("2. Xuất các phần tử thuộc dòng k.")
    print("3. Xuất các phần tử thuộc cột k.")
    print("4. Tìm dòng có tổng lớn nhất.")
    print("5. Tìm cột có tích nhỏ nhất.")
    print("6. Xuất ra các phần tử thuộc dòng chẵn và cột lẻ trong a.")
    print("7. Tính trung bình cộng các phần tử chẵn thuộc dòng lẻ của a.")
    print("8. Tính trung bình cộng các phần tử thuộc biên.")
    print("9. Tính trung bình tích các phần tử không thuộc biên.")
    print("0. Thoát chương trình.")

def process():
    a = []
    m, n = 0, 0
    while True:
        menu()
        try:
            choice = int(input("Nhập vào lựa chọn của bạn: "))
        except ValueError:
            print("Vui lòng nhập một số nguyên!")
            continue

        if choice == 1:
            m = int(input("Nhập vào số dòng của mảng a: "))
            n = int(input("Nhập vào số cột của mảng a: "))
            if m <= 0 or n <= 0:
                print("Số dòng và số cột phải lớn hơn 0.")
                continue

            a = []
            for i in range(m):
                row = []
                for j in range(n):
                    row.append(random.randint(0, 10))
                a.append(row)

            print("Mảng vừa tạo:")
            for i in range(m):
                print(a[i])

        elif not a and choice != 0:
            print("Vui lòng chọn 1 để tạo mảng trước!")
            continue

        elif choice == 2:
            k = int(input(f"Nhập vào dòng k (từ 0 - {m - 1}): "))
            if 0 <= k < m:
                print(f"Các phần tử thuộc dòng {k} là: {a[k]}")
            else:
                print("Dòng k không hợp lệ.")

        elif choice == 3:
            k = int(input(f"Nhập vào cột k (từ 0 - {n - 1}): "))
            if 0 <= k < n:
                print(f"Các phần tử thuộc cột {k} là:")
                for i in range(m):
                    print(a[i][k])
            else:
                print("Cột k không hợp lệ.")

        elif choice == 4:
            max_row_sum(a)
        elif choice == 5:
            min_product_column(a)
        elif choice == 6:
            print("Các phần tử thuộc dòng chẵn và cột lẻ:")
            print_ptu_dchan_cle(a, m, n)
        elif choice == 7:
            avg_ptu_chan_dle(a, m, n)
        elif choice == 8:
            avg_ptu_bien(a, m, n)
        elif choice == 9:
            avg_tich_kbien(a, m, n)
        elif choice == 0:
            print("Đã thoát chương trình.")
            break
        else:
            print("Lựa chọn không hợp lệ. Vui lòng chọn lại!")

process()
```

```python
from datetime import datetime

# a. Xây dựng cấu trúc SINHVIEN mô tả một sinh viên bằng Class
class SinhVien:
    def __init__(self, mssv, ten, nam_sinh, diem_tb):
        self.mssv = mssv
        self.ten = ten
        self.nam_sinh = nam_sinh
        self.diem_tb = diem_tb

# b. Hàm nhập danh sách n sinh viên
def nhap_danh_sach(n):
    ds_sv = []
    for i in range(n):
        print(f"\n--- Nhập sinh viên thứ {i+1} ---")

        while True:
            m = input("Nhập MSSV (10 ký tự): ").strip()
            if len(m) == 10:
                break
            print("Mã sinh viên phải đúng 10 ký tự. Vui lòng nhập lại!")

        while True:
            t = input("Nhập họ và tên (tối đa 20 ký tự): ").strip()
            if 0 < len(t) <= 20:
                break
            print("Tên không được để trống và tối đa 20 ký tự. Vui lòng nhập lại!")

        while True:
            try:
                ns = int(input("Nhập năm sinh: "))
                dtb = float(input("Nhập điểm trung bình: "))
                break # Thoát vòng lặp nếu nhập đúng kiểu số
            except ValueError:
                print("Lỗi: Năm sinh phải là số nguyên và điểm phải là số! Vui lòng nhập lại.")

        # Tạo đối tượng SinhVien và thêm vào danh sách
        ds_sv.append(SinhVien(m, t, ns, dtb))

    return ds_sv

# c. Viết hàm cho biết có bao nhiêu sinh viên đủ điều kiện lên lớp (ĐTB >= 5)
def dem_sv_len_lop(ds_sv):
    count = sum(1 for sv in ds_sv if sv.diem_tb >= 5)
    print(f"\nSố SV đủ điều kiện lên lớp: {count}")

# d. Xuất các sinh viên đủ 20 tuổi
def xuat_sv_du_20_tuoi(ds_sv):
    nam_hien_tai = datetime.now().year # Lấy năm hiện tại tự động (2026)
    print("\nCác SV đủ 20 tuổi trở lên:")
    found = False
    for sv in ds_sv:
        tuoi = nam_hien_tai - sv.nam_sinh
        if tuoi >= 20:
            print(f"- {sv.ten} ({tuoi} tuổi)")
            found = True
    if not found:
        print("- Không có sinh viên nào đủ 20 tuổi.")

# e. Đếm số sinh viên học hệ đại học (MSSV chứa "DH" ở vị trí index 2 và 3)
def dem_sv_he_dai_hoc(ds_sv):
    count_dh = sum(1 for sv in ds_sv if sv.mssv[2:4].upper() == "DH")
    print(f"\nSố SV hệ Đại học: {count_dh}")

# f, g. Cho biết có bao nhiêu SV tên "Lan", họ "Phan"
def thong_ke_ho_ten(ds_sv):
    count_lan = 0
    count_phan = 0

    for sv in ds_sv:
        danh_sach_tu = sv.ten.split()
        if not danh_sach_tu: # Bỏ qua nếu tên trống
            continue

        # Lấy từ cuối cùng (Tên) và từ đầu tiên (Họ)
        if danh_sach_tu[-1].lower() == "lan":
            count_lan += 1
        if danh_sach_tu[0].lower() == "phan":
            count_phan += 1

    print(f"Số SV tên Lan: {count_lan}")
    print(f"Số SV họ Phan: {count_phan}")

if __name__ == "__main__":
    try:
        n = int(input("Nhập số lượng sinh viên: "))
        if n > 0:
            danh_sach = nhap_danh_sach(n)

            # Gọi các hàm xử lý
            dem_sv_len_lop(danh_sach)
            xuat_sv_du_20_tuoi(danh_sach)
            dem_sv_he_dai_hoc(danh_sach)
            thong_ke_ho_ten(danh_sach)
        else:
            print("Số lượng sinh viên phải lớn hơn 0.")
    except ValueError:
        print("Vui lòng nhập một số nguyên hợp lệ cho số lượng sinh viên.")
```
