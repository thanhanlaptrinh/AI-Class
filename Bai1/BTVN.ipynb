```python

#1.	Viết hàm trộn 2 mảng một chiều thành 1 mảng một chiều với mỗi phần tử của mảng mới là min của 2 phần tương ứng từ 2 mảng cho trước. Trong quá trình trộn 2 mảng nếu mảng nào còn phần tử thì các phần tử còn lại của mảng đó sẽ đưa vào mảng mới.
#Ví dụ:
#Mảng a: 3 9 1 4
#Mảng b: 2 7 4 3 2 8
#Mảng kết quả: 2 7 1 3 2 8

def tron_mang_min(a, b):
    mang_ket_qua = []
    min_len = min(len(a), len(b))
    for i in range(min_len):
        mang_ket_qua.append(min(a[i], b[i]))
    if len(a) > len(b):
        mang_ket_qua.extend(a[min_len:])
    else:
        mang_ket_qua.extend(b[min_len:])
    return mang_ket_qua

a = [3, 9, 1, 4]
b = [2, 7, 4, 3, 2, 8]
print("Mảng kết quả:", tron_mang_min(a, b))
```

```python
#2.	Cho mảng 2 chiều a có m dòng, n cột chứa số nguyên, viết các hàm sau:
# a. Tổng tam giác trên đường chéo phụ kể cả đường chéo phụ
def tong_tam_giac_tren_cheo_phu(a, n):
    tong = 0
    for i in range(n):
        for j in range(n):
            if i + j <= n - 1:
                tong += a[i][j]
    return tong

# b. Chuyển phần tử âm thành trị tuyệt đối
def chuyen_tri_tuyet_doi(a):
    for i in range(len(a)):
        for j in range(len(a[0])):
            if a[i][j] < 0:
                a[i][j] = abs(a[i][j])
    return a

# c. Thay các phần tử chẵn bằng x
def thay_chan_bang_x(a, x):
    for i in range(len(a)):
        for j in range(len(a[0])):
            if a[i][j] % 2 == 0:
                a[i][j] = x
    return a

# d. Kiểm tra ma trận có toàn số chẵn không
def kiem_tra_toan_chan(a):
    for row in a:
        for val in row:
            if val % 2 != 0:
                return False
    return True

# e. Kiểm tra ma trận đối xứng (a[i][j] == a[j][i])
def kiem_tra_doi_xung(a, n):
    for i in range(n):
        for j in range(n):
            if a[i][j] != a[j][i]:
                return False
    return True

# f. Kiểm tra đường chéo chính có tăng dần không (a[i][i] < a[i+1][i+1])
def ch_chinh_tang_dan(a, n):
    for i in range(n - 1):
        if a[i][i] >= a[i+1][i+1]:
            return False
    return True

# g. Xuất tam giác dưới đường chéo phụ (kể cả đường chéo phụ: i + j >= n - 1)
def xuat_tam_giac_duoi_cheo_phu(a, n):
    print("Các phần tử tam giác dưới đường chéo phụ:")
    for i in range(n):
        for j in range(n):
            if i + j >= n - 1:
                print(f"a[{i}][{j}]={a[i][j]}", end="  ")
        print()

# h. Kiểm tra đường chéo phụ có giảm dần không (a[i][n-1-i] > a[i+1][n-2-i])
def ch_phu_giam_dan(a, n):
    for i in range(n - 1):
        if a[i][n - 1 - i] <= a[i + 1][n - 2 - i]:
            return False
    return True
```
