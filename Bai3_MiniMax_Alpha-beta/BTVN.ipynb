```python
#Bài 3: Cài đặt thuật toán minimax cho bài toán tictactoe với ma trận 5x5, 10x10, NxN.

import copy
import math
import numpy as np
import os

N = 5
WIN_STREAK = 4

X = "X"
O = "O"
EMPTY = None

def initial_state():
    return [[EMPTY for _ in range(N)] for _ in range(N)]

def player(board):
    count = sum(1 for row in board for cell in row if cell != EMPTY)
    return O if count % 2 != 0 else X

def actions(board):
    res = set()
    for i in range(N):
        for j in range(N):
            if board[i][j] == EMPTY:
                res.add((i, j))
    return res

def result(board, action):
    curr_player = player(board)
    result_board = copy.deepcopy(board)
    i, j = action
    result_board[i][j] = curr_player
    return result_board

def winner(board):
    for i in range(N):
        for j in range(N):
            curr = board[i][j]
            if curr == EMPTY: continue

            if j + WIN_STREAK <= N and all(board[i][j+k] == curr for k in range(WIN_STREAK)):
              return curr
            if i + WIN_STREAK <= N and all(board[i+k][j] == curr for k in range(WIN_STREAK)):
              return curr
            if i + WIN_STREAK <= N and j + WIN_STREAK <= N and all(board[i+k][j+k] == curr for k in range(WIN_STREAK)):
              return curr
            if i + WIN_STREAK <= N and j - WIN_STREAK >= -1 and all(board[i+k][j-k] == curr for k in range(WIN_STREAK)):
              return curr
    return None

def terminal(board):
    if winner(board) is not None:
        return True
    return len(actions(board)) == 0

def utility(board):
    win_val = winner(board)
    if win_val == X:
      return 1000
    elif win_val == O:
      return -1000
    return 0

def heuristic(board):
    return utility(board)

def maxValue(state, depth):
    if terminal(state):
        return utility(state)
    if depth == 0:
        return heuristic(state)

    v = -math.inf
    for action in actions(state):
        v = max(v, minValue(result(state, action), depth - 1))
    return v

def minValue(state, depth):
    if terminal(state):
        return utility(state)
    if depth == 0:
        return heuristic(state)

    v = math.inf
    for action in actions(state):
        v = min(v, maxValue(result(state, action), depth - 1))
    return v

def minimax(board, max_depth=3):
    current_player = player(board)
    best_move = None

    if current_player == X:
        max_val = -math.inf
        for action in actions(board):
            check = minValue(result(board, action), max_depth - 1)
            if check > max_val:
                max_val = check
                best_move = action
    else:
        min_val = math.inf
        for action in actions(board):
            check = maxValue(result(board, action), max_depth - 1)
            if check < min_val:
                min_val = check
                best_move = action
    return best_move

def print_board(board):
    os.system('cls' if os.name == 'nt' else 'clear')
    display_board = np.where(np.array(board) == None, "-", np.array(board))
    print(display_board)
    print("-" * 20)

if __name__ == "__main__":
    board = initial_state()
    user = input("Chọn người chơi (X/O): ").strip().upper()
    ai = O if user == X else X

    while True:
        print_board(board)
        if terminal(board):
            win_player = winner(board)
            print("Game Over: Hòa." if win_player is None else f"Game Over: {win_player} thắng.")
            break

        curr_player = player(board)
        if curr_player == user:
            print("Nhập tọa độ nước đi (Ví dụ: 0 1):")
            try:
                i, j = map(int, input("Row Col: ").split())
                if board[i][j] == EMPTY:
                    board = result(board, (i, j))
                else:
                    print("Ô đã có người đánh!")
            except:
                print("Lỗi nhập liệu!")
        else:
            print(f"AI ({ai}) đang tính toán (độ sâu 3)...")
            move = minimax(board, max_depth=3)
            if move:
                board = result(board, move)
```

```python
#Bài 4. Cài đặt thuật toán cắt tỉa Alpha-beta cho bài toán tictactoe với ma trận 5x5, 10x10, NxN.
import copy
import math
import os

N = 5
WIN_STREAK = 4
MAX_DEPTH = 4

X = "X"
O = "O"
EMPTY = None

def initial_state():
    return [[EMPTY for _ in range(N)] for _ in range(N)]

def player(board):
    count = sum(1 for row in board for cell in row if cell != EMPTY)
    return O if count % 2 != 0 else X

def actions(board):
    return [(i, j) for i in range(N) for j in range(N) if board[i][j] == EMPTY]

def result(board, action):
    res_board = copy.deepcopy(board)
    res_board[action[0]][action[1]] = player(board)
    return res_board

def get_winner(board):
    for i in range(N):
        for j in range(N):
            curr = board[i][j]
            if curr == EMPTY: continue
            if j + WIN_STREAK <= N and all(board[i][j+k] == curr for k in range(WIN_STREAK)):
              return curr
            if i + WIN_STREAK <= N and all(board[i+k][j] == curr for k in range(WIN_STREAK)):
              return curr
            if i + WIN_STREAK <= N and j + WIN_STREAK <= N and all(board[i+k][j+k] == curr for k in range(WIN_STREAK)):
              return curr
            if i + WIN_STREAK <= N and j - WIN_STREAK >= -1 and all(board[i+k][j-k] == curr for k in range(WIN_STREAK)):
              return curr
    return None

def minimax_alpha_beta(board, depth, alpha, beta, is_maximizing):
    win_val = get_winner(board)
    if win_val == X: return 1000 + depth
    if win_val == O: return -1000 - depth

    available_actions = actions(board)
    if len(available_actions) == 0 or depth == 0:
        return 0

    if is_maximizing:
        max_eval = -math.inf
        for action in available_actions:
            eval = minimax_alpha_beta(result(board, action), depth - 1, alpha, beta, False)
            max_eval = max(max_eval, eval)
            alpha = max(alpha, eval)
            if beta <= alpha: break
        return max_eval
    else:
        min_eval = math.inf
        for action in available_actions:
            eval = minimax_alpha_beta(result(board, action), depth - 1, alpha, beta, True)
            min_eval = min(min_eval, eval)
            beta = min(beta, eval)
            if beta <= alpha: break
        return min_eval

def find_best_move(board):
    ai_player = player(board)
    best_move = None

    if ai_player == X:
        best_val = -math.inf
        for action in actions(board):
            move_val = minimax_alpha_beta(result(board, action), MAX_DEPTH - 1, -math.inf, math.inf, False)
            if move_val > best_val:
                best_val = move_val
                best_move = action
    else:
        best_val = math.inf
        for action in actions(board):
            move_val = minimax_alpha_beta(result(board, action), MAX_DEPTH - 1, -math.inf, math.inf, True)
            if move_val < best_val:
                best_val = move_val
                best_move = action
    return best_move

def print_board(board):
    os.system('cls' if os.name == 'nt' else 'clear')
    print(f"\n--- BÀN CỜ {N}x{N} ---")
    for i in range(N):
        row_display = []
        for j in range(N):
            if board[i][j] == EMPTY:
                idx = i * N + j + 1
                row_display.append(f"{idx:2}")
            else:
                row_display.append(f"{board[i][j]:>2}")

        print(" | ".join(row_display))
        if i < N - 1:
            print("-" * (N * 5 - 1))
    print()

if __name__ == "__main__":
    board = initial_state()
    user = input("Bạn muốn chơi X hay O? (X đi trước): ").strip().upper()
    ai = O if user == X else X

    while True:
        print_board(board)
        winner_val = get_winner(board)

        if winner_val is not None:
            print(f"TRÒ CHƠI KẾT THÚC: Người chơi '{winner_val}' thắng!")
            break
        if len(actions(board)) == 0:
            print("TRÒ CHƠI KẾT THÚC: Bàn cờ đã đầy. Hai bên hòa nhau.")
            break

        curr_player = player(board)
        if curr_player == user:
            while True:
                try:
                    num = int(input(f"Đến lượt bạn ({user}). Chọn số từ 1 đến {N*N}: "))
                    if 1 <= num <= N*N:
                        i = (num - 1) // N
                        j = (num - 1) % N

                        if board[i][j] == EMPTY:
                            board = result(board, (i, j))
                            break
                        else:
                            print("Lỗi: Ô này đã có người đánh! Vui lòng chọn ô khác.")
                    else:
                        print(f"Lỗi: Số nhập vào phải nằm trong khoảng từ 1 đến {N*N}.")
                except ValueError:
                    print("Lỗi: Vui lòng nhập một số nguyên hợp lệ!")
        else:
            print(f"AI ({ai}) đang suy nghĩ...")
            move = find_best_move(board)
            if move:
                board = result(board, move)
```

```python
#Bài 6. Xây dựng giao diện đồ họa cho trò chơi tictactoe.
import tkinter as tk
from tkinter import messagebox
import math

X = "X"
O = "O"
EMPTY = ""

class TicTacToe:
    def __init__(self, root):
        self.root = root
        self.root.title("Tic Tac Toe AI (Alpha-Beta)")
        self.board = [[EMPTY for _ in range(3)] for _ in range(3)]
        self.buttons = [[None for _ in range(3)] for _ in range(3)]
        self.human = X
        self.ai = O

        self.create_gui()

    def create_gui(self):
        for i in range(3):
            for j in range(3):
                btn = tk.Button(self.root, text=EMPTY, font=('Arial', 40, 'bold'),
                                width=5, height=2, bg="#f0f0f0",
                                command=lambda row=i, col=j: self.human_move(row, col))
                btn.grid(row=i, column=j, padx=2, pady=2)
                self.buttons[i][j] = btn

        reset_btn = tk.Button(self.root, text="Chơi lại", font=('Arial', 14), command=self.reset_game)
        reset_btn.grid(row=3, column=0, columnspan=3, sticky="we", pady=5)

    def get_winner(self, board):
        for i in range(3):
            if board[i][0] == board[i][1] == board[i][2] != EMPTY: return board[i][0]
            if board[0][i] == board[1][i] == board[2][i] != EMPTY: return board[0][i]
        if board[0][0] == board[1][1] == board[2][2] != EMPTY: return board[0][0]
        if board[0][2] == board[1][1] == board[2][0] != EMPTY: return board[0][2]
        return None

    def is_draw(self, board):
        for row in board:
            if EMPTY in row: return False
        return True

    def human_move(self, row, col):
        if self.board[row][col] == EMPTY and self.get_winner(self.board) is None:
            self.board[row][col] = self.human
            self.buttons[row][col].config(text=self.human, fg="blue")

            if self.check_game_over(): return

            self.root.update()
            self.ai_move()

    def ai_move(self):
        best_score = -math.inf
        best_move = None

        for i in range(3):
            for j in range(3):
                if self.board[i][j] == EMPTY:
                    self.board[i][j] = self.ai
                    score = self.alpha_beta(self.board, 0, -math.inf, math.inf, False)
                    self.board[i][j] = EMPTY
                    if score > best_score:
                        best_score = score
                        best_move = (i, j)

        if best_move:
            self.board[best_move[0]][best_move[1]] = self.ai
            self.buttons[best_move[0]][best_move[1]].config(text=self.ai, fg="red")
            self.check_game_over()

    def alpha_beta(self, board, depth, alpha, beta, is_maximizing):
        winner = self.get_winner(board)
        if winner == self.ai: return 10 - depth
        if winner == self.human: return -10 + depth
        if self.is_draw(board): return 0

        if is_maximizing:
            max_eval = -math.inf
            for i in range(3):
                for j in range(3):
                    if board[i][j] == EMPTY:
                        board[i][j] = self.ai
                        eval = self.alpha_beta(board, depth + 1, alpha, beta, False)
                        board[i][j] = EMPTY
                        max_eval = max(max_eval, eval)
                        alpha = max(alpha, eval)
                        if beta <= alpha: break
            return max_eval
        else:
            min_eval = math.inf
            for i in range(3):
                for j in range(3):
                    if board[i][j] == EMPTY:
                        board[i][j] = self.human
                        eval = self.alpha_beta(board, depth + 1, alpha, beta, True)
                        board[i][j] = EMPTY
                        min_eval = min(min_eval, eval)
                        beta = min(beta, eval)
                        if beta <= alpha: break
            return min_eval

    def check_game_over(self):
        winner = self.get_winner(self.board)
        if winner:
            messagebox.showinfo("Kết thúc", f"Người chơi {winner} chiến thắng!")
            return True
        if self.is_draw(self.board):
            messagebox.showinfo("Kết thúc", "Hòa nhau rồi!")
            return True
        return False

    def reset_game(self):
        self.board = [[EMPTY for _ in range(3)] for _ in range(3)]
        for i in range(3):
            for j in range(3):
                self.buttons[i][j].config(text=EMPTY, state=tk.NORMAL)

if __name__ == "__main__":
    root = tk.Tk()
    app = TicTacToe(root)
    root.eval('tk::PlaceWindow . center')
    root.mainloop()
```
