```python
#Bài tập về nhà
import random

TEACHERS = ['Thầy An', 'Cô Bình', 'Thầy Chi', 'Cô Diệp']
SLOTS = 10

def get_fitness(individual):
    conflicts = 0
    for t in TEACHERS:
        if individual.count(t) > 3:
            conflicts += 1
    return 1 / (1 + conflicts)

def genetic_algorithm():
    pop_size = 10
    population = [[random.choice(TEACHERS) for _ in range(SLOTS)] for _ in range(pop_size)]

    for gen in range(50):
        population.sort(key=lambda ind: get_fitness(ind), reverse=True)

        parent1, parent2 = population[0], population[1]
        point = SLOTS // 2
        child = parent1[:point] + parent2[point:]

        if random.random() < 0.1:
            child[random.randint(0, SLOTS-1)] = random.choice(TEACHERS)

        population[-1] = child

    return population[0]

best_schedule = genetic_algorithm()
print("--- KẾT QUẢ SẮP LỊCH (GA) ---")
for i, t in enumerate(best_schedule):
    print(f"Tiết {i+1}: {t}")
```
